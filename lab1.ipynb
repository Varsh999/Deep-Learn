```python
import numpy as np

# -------------------------------
# Inputs and initial values
# -------------------------------
inputs = np.array([1.0, 2.0, 3.0])
initial_weights = np.array([1.0, 0.5, 0.5])
initial_bias = 0.0

learning_rate = 0.1
target = 1.0


# -------------------------------
# Function: Forward Pass
# -------------------------------
def forward_pass(inputs, weights, bias):
    return np.dot(weights, inputs) + bias


# -------------------------------
# Case 1: Fixed Bias (b = 0)
# -------------------------------
print("----- Case 1: Fixed Bias (b = 0) -----")

weights_fixed = initial_weights.copy()
bias_fixed = 0.0

output_fixed = forward_pass(inputs, weights_fixed, bias_fixed)

print("Output:", output_fixed)
print("Weights (no update):", weights_fixed)
print("Bias (fixed):", bias_fixed)


# -------------------------------
# Case 2: Varying Bias (update)
# -------------------------------
print("\n----- Case 2: Varying Bias (updated) -----")

weights_var = initial_weights.copy()
bias_var = initial_bias

# Forward pass
output_var = forward_pass(inputs, weights_var, bias_var)

# Error
error = target - output_var

# Update weights (vectorized)
weights_var = weights_var + learning_rate * error * inputs

# Update bias
bias_var = bias_var + learning_rate * error

# Print results
print("Output before update:", output_var)
print("Updated Weights:", weights_var)
print("Updated Bias:", bias_var)
```
